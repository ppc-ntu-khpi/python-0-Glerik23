1. Обчисліть радіус кола, знаючи його площу. Нагадаємо формулу:
$$
  S=\pi*R^2 => R =\sqrt{\frac{S}{\pi}}
$$
Для обчислення скористаємось знайомими функціями ```sqrt``` та ```pi``` з пакету ```math```.


````python
from math import sqrt,pi
S = float(input('Введіть площу кола: '))
r = sqrt(S/pi)
print ('Радіус дорівнює %6.3f см' %r)
````

2. А тепер за зразком знайдіть в наступній комірці площу сфери та об'єм кулі. Нагадуємо формули:
$$
  S=4*\pi*R^2    
$$
$$
  V=\frac{4}{3}*\pi*R^3
$$
Також нагадаємо, що ступінь в Python задається так: ```**```

````python
from math import pi

R = float(input('Введiть радiус: '))

# Площа сфери
S = 4 * pi * R**2
print(f'Площа сфери = {S}')

# Об'єм кулі
V = (4/3) * pi * R**3
print(f'Об\'єм кулi = {V}')
````

3. А ще давайте переведемо температуру за Фаренгейтом у градуси Цельсія і навпаки. Формули тут надзвичайно прості:
$$
  C=(F-32)*0.5556
$$
$$
  F=C*1.8+32
$$
Спробуйте в наступній комірці! Пам'ятайте про "красиве" введення і виведення даних.

````python
from math import pi
# Введення температури в Фаренгейтах та Цельсіях
F = float(input('Введiть температуру за Фаренгейтом: '))
C = float(input('Введiть температуру за Цельсiем: '))

# Перетворення з Фаренгейта в Цельсій
tempC = round((F - 32) * 0.5556, 1)

# Перетворення з Цельсія у Фаренгейт
tempF = round(C * 1.8 + 32, 1)

# Виведення результатів
print(f'Температура за Цельсiем: {tempC}')
print(f'Температура за Фаренгейтом: {tempF}')
````
